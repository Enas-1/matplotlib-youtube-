```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
plt.figure(figsize=(8,5))
# add font (fontdict {})
plt.title('Gas price over time in USA',fontdict={'fontweight':'bold','fontsize':18})


plt.plot(gas.Year, gas.USA, 'b.-',label='United States')
plt.plot(gas.Year , gas.Canada, 'g.-',label='Canada')
# anther way to show column within []
plt.plot(gas['Year'],gas['South Korea'],'r.-', label='South Korea')
plt.plot(gas.Year , gas.Australia,'y.-',label='Australia')

# a way to show years arrenged (skip three years ),,, print(gas.Year[::3])
# add anither year to the end of the years ,,,, .tolist()+[].
plt.xticks(gas.Year[::3].tolist()+[2011])
# label = legend

plt.xlabel('Year')
plt.ylabel('US dollers')

plt.legend()

# Saving our graph (the higher the number the higher the resoultion)
plt.savefig('Gas_price_figure.png',dpi=300)

plt.show()
```

```python
for country in gas: 
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker='.')

plt.show()
```

```python
# another way to plot specific countries (many values):
countries_to_look_at=['Australia','USA','Canada','South Korea']
for country in countries_to_look_at:
    plt.plot(gas.Year , gas[country], marker='.')
```
